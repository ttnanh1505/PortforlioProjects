```python
fig = plt.figure(figsize = (12,6))
plt.plot(df_bitcoin['date'], df_bitcoin['open'], color = 'red')
plt.plot(df_bitcoin['date'], df_bitcoin['close'], color = 'blue')
plt.title('Open and Close price of bitcoin over the years')
plt.legend(['open', 'close'])
plt.xlabel('date')
plt.ylabel('price')
```

```python
df_bitcoin['year'] = df_bitcoin['date'].astype('string').str[0:4]
```

```python
df_bitcoin.head()
```

```python
bitcoin_pivot = pd.pivot_table(df_bitcoin, index = 'year', aggfunc = ['mean'])
bitcoin_pivot
```

```python
# bitcoin growth rate over the years
bitcoin_growth = []
a = 0
for i in range(0,7):
    a = (bitcoin_pivot['mean']['close'][i+1])*100 / (bitcoin_pivot['mean']['close'][i]) - 100
    bitcoin_growth.append(a)
bitcoin_growth
```

```python
# average bitcoin growth rate
sum(bitcoin_growth)/len(bitcoin_growth)
```

```python
sp_growth = []
a = 0
for i in range(0,7):
    a = 100*pivot_500['close'][i+1] / pivot_500['close'][i] - 100
    sp_growth.append(a)
sp_growth
```
